## Vorbereitung: Materialien

Für alle folgenden Aufgaben wollen wir die verfügbaren Materialien wissen, die der Server kennt. Also fragen wir ihn:

```python
for node_type in mt.node.type:
    print(node_type)
```

... das ergibt eine sehr lange Liste, die das Jupyter-Notebook für uns netterweise abschneidet. Man kann sie vollständig ausklappen, um etwas gewünschtes zu suchen. Man kann jedoch auch einen oder mehrere Such-Texte verwenden, um die Einträge zu filtern:

```python
must_contain= [ "wool"]
must_not_contain= ["slab", "slope"]

for n in mt.node.type:
    match= True
    for f in must_contain:
        if f not in n:
            match= False
            break
    for f in must_not_contain:
        if f in n:
            match= False
            break
    if match:
        print(n)
```

... damit können wir uns schrittweise zu gewünschten Materialien vortasten.

## Aufgabe: Ebenen aus Blöcken

Es soll eine Ebene aus Blöcken entlang der Koordinaten-Richtungen angelegt werden. Das kann man in x-z-Richtung tun, also horizontal, oder in x-y-Richtung oder y-z-Richtung vertikal, also als Wand. Man muss jeweils 2 Längen angeben, die je nach Fall Höhe, Breite oder Länge festlegen.

### Teilaufgabe 03a: Block für Block horizontal

Programmiere zwei in einander geschachtelte Schleifen, die Blöcke in x- und z- Richtung setzen, mit `w` Blöcken in x-Richtung und `d` Blöcken in z-Richtung. Die Blöcke sollen schritt für Schritt in der Landschaft erscheinen.

### Teillösung 03a

```python
# get current position of player
pos= mtb.pos_as_int(player)

# set width and length
w= 5
d= 9

for x in range(w):
    for z in range(d):
        mtb.set(mt, pos, [[x,0,z]], "wool:yellow")
```

### Teilaufgabe 03b: Alle Blöcke auf einmal

Programmiere eine weitere Variante, die alle Blöcke auf einen Schlag erzeugt. Sammle dafür die Koordinatenvektroren in einer Liste und übergib sie am Ende an `mtb.set()`.

### Teillösung 03b

```python
# get current position of player
pos= mtb.pos_as_int(player)

# set width and length
w= 5
d= 9

positions= []
for x in range(w):
    for z in range(d):
        positions.append([x,0,z])

mtb.set(mt, pos, positions, "wool:orange")
```

### Teilaufgabe 03c: Funktionen

Programmiere eine Funktion (auch „Routine“ oder „Subroutine“ genannt), die mit den benötigten Parametern eine Ebene von Blöcken mit einem gegebenen Material erstellt. Dieses Mal als vertikale Wand in x-y-Richtung. Sie benötigt das mt-Objekt, die Position der Spielfigur, sowie die Koordinaten und das Material. Dabei können die Koordinaten als Startpunkt `s` (3D-Vektor) zusammen mit Länge `l` und Höhe `h` angegeben werden.

Am Ende rufe die Funktion auf um eine Wand neben der Spielfigur erstellen (daneben aber nicht an der Stelle der Spielfigur)

### Teillösung 03c

```python
def wall( mt, pos, startpos, width, height, material ):

    positions= []
    for x in range(width):
        for y in range(height):
            positions.append([x,y,0])

    mtb.set(mt, pos+startpos, positions, material)

# get current position of player
pos= mtb.pos_as_int(player)

wall( mt, pos, [-5,1,2], 10, 5, "wool:cyan" )
```
